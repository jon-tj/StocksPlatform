```python
url = "https://www.nordnet.no/api/2/instrument_search/query/bullbearlist?apply_filters=underlying_name%3DEQUINOR%7Cnordnet_markets%3Dtrue&limit=100"
headers = {
    "accept": "application/json",
    "ntag": "NO_NTAG_RECEIVED_YET",
    "client-id": "NEXT",
  }

response = requests.get(url, headers=headers)
data = response.json()

certificates = [
    {
        "holders": r["statistical_info"]["number_of_owners"],
        "gearing": r["certificate_info"]["static_leverage"],
        "direction": r["etp_info"]["direction"],
     }
     for r in data["results"]
     ]

certificates

ratio_long = sum(c["holders"] ** 0.7 * c["gearing"] for c in certificates if c["direction"] == "Long") / sum(c["holders"] ** 0.7 * c["gearing"] for c in certificates)
bnb_delta = ratio_long * 2 - 1
bnb_delta
```
